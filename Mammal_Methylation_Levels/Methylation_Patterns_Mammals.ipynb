### Analysis of Methylation Data from the Mammalian Methylation Consortium dataset (Horvath 2023)

```python
# Import 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
methylation_data = pd.read_csv('GSE223748_datBetaNormalized.csv')
```

```python
print(methylation_data.shape)
print(methylation_data.head(1))

# 37554 CpG sites and 15044 samples
#(37554, 15044)
```

```python
methylation_data.head(1)
```

```python
# Rename column Unnamed: 0 to CpG
methylation_data.rename(columns={'Unnamed: 0': 'CpG'}, inplace=True)
```

```python
# Check if there are any missing values
missing_values = methylation_data.isnull().sum().sum()
print(missing_values)
```

```python
# Check if there are any duplicated CpG sites
duplicated_cpg = methylation_data['CpG'].duplicated().sum()
print(duplicated_cpg)
```

```python
# Summary statistics
summary_stats = methylation_data.describe()
print(summary_stats)
```

```python
# Take a sample of 1000 CpG sites (keep the first column has it has the CpG site names)
methylation_data_sample = methylation_data.iloc[:,0:1001]
```

```python
methylation_data_sample.shape
```

```python
methylation_data_sample.head(1)
```

```python
# Exclude the first column (CpG names) from the dataframe before plotting
methylation_data_sample_without_names = methylation_data_sample.iloc[:, 1:]

plt.figure(figsize=(10, 8))
sns.heatmap(methylation_data_sample_without_names, cmap='viridis')
plt.xlabel('Samples')
plt.ylabel('CpG Sites')
plt.title('Methylation Patterns')
plt.show()
```
